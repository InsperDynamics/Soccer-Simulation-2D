```python
import os
import numpy as np
import pandas as pd
import random
```

```python
rcg_files = os.listdir('rcg')
rcl_files = os.listdir('rcl')
invalid_rcg_files = []

for rcg in rcg_files:
    if not rcg.replace('rcg', 'rcl') in rcl_files:
        invalid_rcg_files.append(rcg)
    elif rcg.startswith('2008040') or rcg.startswith('200803'):
        invalid_rcg_files.append(rcg)

for invalid_rcg in invalid_rcg_files:
    rcg_files.remove(invalid_rcg)
```

```python
print('Number of valid games: {}'.format(len(rcg_files)))
print('Number of valid observations: {}'.format(len(rcg_files) * 6000))
```

```python
random.shuffle(rcg_files)
print("Shuffled: " + str(rcg_files[0:10]))
```
